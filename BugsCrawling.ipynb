```python
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen
import pandas 
import selenium 
```

```python
url = "https://music.bugs.co.kr/chart"
url = requests.get(url)
soup = BeautifulSoup(url.content, "html.parser")
```

```python
soup
```

```python
body = soup.find_all('tbody')
```

```python
array = body[0]
type(array)
```

```python
print(array)
```

```python
musiclist = array.find_all('tr')
```

```python
len(musiclist)
```

```python
musiclist[0].find_all('p', class_ = 'title')[0].get_text().strip()
```

```python
musiclist[0].find_all('p', class_ = 'artist')[0].get_text().strip()
```

```python
musiclist[0].find_all('a', class_ = 'album')[0].get_text().strip()
```

```python
title_list = []
artist_list = []
album_list = []
count = len(musiclist)
for i in range(count):
    title = musiclist[i].find_all('p', class_ = 'title')[0].get_text().strip()
    artist = musiclist[i].find_all('p', class_ = 'artist')[0].get_text().strip()
    album = musiclist[i].find_all('a', class_ = 'album')[0].get_text().strip()
    print(title,"", artist, "", album, "")
    title_list.append(title)
    artist_list.append(artist)
    album_list.append(album)
    
```

```python
import pandas as pd
pd.set_option('display.max_row', 500)
pd.set_option('display.max_columns', 100)
music = pd.DataFrame({"title":title_list,"artist":artist_list,"album":album_list})
```
