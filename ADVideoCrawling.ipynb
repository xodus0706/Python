# 김태연 게임 동물 영화애니메이션 인물블로그 코미디

```python
data = pd.read_csv('C:/Users/Home/Desktop/인물블로그광고.csv' , sep=",", encoding='cp949') #utf8 이 오류나면 cp949 반대면 반대
                 ###드라이브에 1차 가공 안에서 각자 맡은 카테고리csv를 저장한 경로
urlList = data['광고동영상url']
header = "https://www.youtube.com/watch?v="
count = len(urlList)
```

```python
for i in tqdm(range(274,count)):
    geturl = header + urlList[i]
    try:
        driver = webdriver.Chrome(executable_path='C:/Users/Home/chromedriver') 
        # 드라이브 경로 한번만 바꾸셈!
        driver.get(url=geturl)
        time.sleep(2.0) #값 조정 꼭!!!
        for j in range(2):
            driver.execute_script("window.scrollTo(0, 400);") 
            time.sleep(3.0) #값 조정 꼭!!!
        click = driver.find_elements_by_tag_name('div > div > div > ytd-video-view-count-renderer > span')[0].text
        like = driver.find_elements_by_tag_name('div > div > div > ytd-menu-renderer > div > ytd-toggle-button-renderer > a > yt-formatted-string')[2].text
        dislike = driver.find_elements_by_tag_name('div > div > div > ytd-menu-renderer > div > ytd-toggle-button-renderer > a > yt-formatted-string')[3].text
        comment = driver.find_elements_by_tag_name('ytd-comments-header-renderer > div > h2 > yt-formatted-string > span')[1].text
        data.loc[i,'조회수'] = click
        data.loc[i,'좋아요'] = like
        data.loc[i,'싫어요'] = dislike
        data.loc[i,'댓글수'] = comment
    except:
        print(i," 번째 행 안됐음 나중에 확인하시길 바람...")
    time.sleep(2.0) #값 조정 꼭!1
    driver.quit()
```
