```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium.webdriver.common.keys import Keys
import time
from tqdm.notebook import tqdm
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
header = "https://playboard.co/channel/"
data = pd.read_csv('C:/Users/Home/Downloads/9.교육.csv' , sep=",", encoding='cp949')
# 드라이브의 03.수집 -> 채널테이블 폴더에서 다운 받은 경로를 입력
# 본인이 맡은 카테고리 경로와 폴더이름 입력
primList = data['주키']
nameList = data['채널 이름']
footer = "/videos"
count = len(primList)
adurl_list = []
channelurl_list = []
name_list = []
day_list = []
cate_list = []
```

```python
cateList = [   "코미디","엔터테이먼트","자동차","게임",  #0 1 2 3   박영상
               "스포츠","뉴스정치","노하우스타일",       #4 5 6     박영상
               "여행이벤트","음악",                      #7 8   정광의
                "동물","교육",                           #9 10  김태연
               "인물블로그","영화애니메이션"]            #11 12 서예지
```

```python
for i in range(7,8):
    geturl = header + primList[i] + footer
    try:
        driver = webdriver.Chrome(executable_path='C:/Users/Home/chromedriver') 
        # 드라이브 경로 한번만 바꾸셈!
        driver.get(url=geturl)
        time.sleep(2.0)
        adclick = driver.find_elements_by_class_name('txt')
        adclick[3].click()
        flag = 0
        isExited = False
        prev_height = driver.execute_script("return document.body.scrollHeight")
        time.sleep(2.0)
        while(True):
            dayList = driver.find_elements_by_class_name('date')
            if(len(dayList) == 0):
                break
            for j in range(len(dayList)):
                day = dayList[j].text
                lastday = dayList[len(dayList)-1].text
                if '2021' not in day:
                    flag = j
                    isExited = True
                    break
                elif '2021' in lastday:
                    break
            if(isExited):
                break
            driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
            time.sleep(5.0)
            curr_height = driver.execute_script("return document.body.scrollHeight")
            if(curr_height == prev_height):
                flag = len(dayList)-1
                break
            else:
                prev_height = driver.execute_script("return document.body.scrollHeight")
        
        dayList = driver.find_elements_by_class_name('date')
        urlList = driver.find_elements_by_tag_name('div > div > div > article > div > a')
        for j in tqdm(range(0,flag+1)):
            uuu = urlList[j].get_attribute('href')
            adurl = uuu[27:]
            adurl_list.append(adurl)
            channelurl_list.append(primList[i])
            name_list.append(nameList[i])
            day_list.append(dayList[j].text)
            cate_list.append(cateList[10])  # 본인이 맡은 카테고리 인덱스 번호 입력
    except:
        print(nameList[i]," <-이 채널에 유료광고 없는거임 ㅇㅋ? ",(i+1)," 번째 안됨")
    time.sleep(1.0)
    driver.quit()
```

```python
df = pd.DataFrame({"광고동영상url":adurl_list,"채널url":channelurl_list,
                   "채널이름":name_list,"업로드날짜":day_list,
                  "카테고리":cate_list,"좋아요":"","싫어요":"",
                   "비율":"","조회수":"",
                   "댓글수":"","언급도":""})
```
