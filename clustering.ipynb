```python
import mglearn
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc
```

```python
font_location = "c:/Windows/fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
matplotlib.rc("font", family = font_name)
```

```python
data = pd.read_csv("C:/Users/Home/Documents/ad_46_1.csv")
kmeans = KMeans(n_clusters = 2)
kmeans.fit(data.iloc[:,19:]) #adsumview기준
print("클러스터 레이블", kmeans.labels_)
```

```python
mglearn.discrete_scatter(data.iloc[:,19], data.iloc[:,8], kmeans.labels_)
plt.legend(["클러스터0", "클러스터1"], loc="best")
plt.xlabel("adsumview")
plt.ylabel("adcount")
plt.show()
```
